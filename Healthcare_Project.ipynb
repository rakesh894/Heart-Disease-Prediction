# Prepare Data for modeling

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

## Model 1 : Logistic Regression

```python
from sklearn.metrics import classification_report 
from sklearn.linear_model import LogisticRegression

M1 = LogisticRegression(random_state=1) # get instance of model
M1.fit(x_train, y_train) # Train/Fit model 

y_hat = M1.predict(x_test) # get y predictions
print(classification_report(y_test, y_hat)) # output accuracy


```

##  K-NN( K- Nearest Neighbors)

```python
from sklearn.metrics import classification_report 
from sklearn.neighbors import KNeighborsClassifier

M2 = KNeighborsClassifier() # get instance of model
M2.fit(x_train, y_train) # Train/Fit model 

y_hat2 = M2.predict(x_test) # get y predictions
print(classification_report(y_test, y_hat2)) # output accuracy
```

## Model 3 : SVM( Support Vector Machine)

```python
from sklearn.metrics import classification_report 
from sklearn.svm import SVC

svc_model = SVC(random_state=1) # get instance of model
svc_model.fit(x_train, y_train) # Train/Fit model 

y_hat3 = svc_model.predict(x_test) # get y predictions
print(classification_report(y_test, y_hat3)) # output accuracy

print(svc_model.score(x_train, y_train))
print(svc_model.score(x_test, y_test))
```

## Model 3:  Naive Bayes Classifier

```python
from sklearn.metrics import classification_report 
from sklearn.naive_bayes import GaussianNB

M4 = GaussianNB() # get instance of model
M4.fit(x_train, y_train) # Train/Fit model 

y_hat4 = M4.predict(x_test) # get y predictions
print(classification_report(y_test, y_hat4)) # output accuracy
```

##  Decision Trees

```python
from sklearn.metrics import classification_report 
from sklearn.tree import DecisionTreeClassifier

M5 = DecisionTreeClassifier(random_state=1) # get instance of model
M5.fit(x_train, y_train) # Train/Fit model 

y_hat5 = M5.predict(x_test) # get y predictions
print(classification_report(y_test, y_hat5)) # output accuracy
```

## Model 4: Random Forest 

```python
from sklearn.metrics import classification_report 
from sklearn.ensemble import RandomForestClassifier

model6 = RandomForestClassifier(random_state=1)# get instance of model
model6.fit(x_train, y_train) # Train/Fit model 

y_pred6 = model6.predict(x_test) # get y predictions
print(classification_report(y_test, y_pred6)) # output accuracy
```

```python
from numpy import unique
from numpy import argmax
from pandas import read_csv
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import plot_model
```

```python
n_features = X.shape[1]
n_class = len(unique(y))
```

```python
from pandas import read_csv
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from keras.layers import Dropout
```

```python
network7 = Sequential()

network7.add(Dense(20, input_dim=n_features, activation='relu',kernel_initializer='RandomNormal'))
network7.add(Dense(30,activation='relu'))
network7.add(Dense(15,activation='relu'))
network7.add(Dense(1,activation='softmax'))

network7.summary()


opt = keras.optimizers.Adam(learning_rate=0.0009)

network7.compile(optimizer=opt,
                loss='categorical_crossentropy',
                metrics=['accuracy'])


network7.fit(x_train, y_train, epochs=20, batch_size=32, verbose=1)

test_loss_7, test_acc_7 = network7.evaluate(x_test, y_test)
print('test_acc_7:', test_acc_7)
```
